# Get Group Method

```python
data.head(3)
```

```python
data['COUNTRY'].nunique()
```

```python
country_usa = data[data['COUNTRY'] == 'USA']
country_france = data[data['COUNTRY'] == 'France']
country_india = data[data['COUNTRY'] == 'India']
```

```python
group_country = data.groupby('COUNTRY')
```

# Sum, Mean, Max, Min Method

# .agg method

```python
data.head(3)
```

```python
group_country.agg('sum')
```

```python
group_country.agg('sum')[['QUANTITYORDERED','SALES']]
```

```python
group_country.agg(['sum','mean'])[['QUANTITYORDERED','SALES']]
```
